```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(42)
```

```python
X_train,X_test,y_train,y_test=train_test_split(DF.iloc[:,:-1],DF.iloc[:,-1],test_size=0.3,random_state=42) # Splitting the given data in 30% test and 70% train
```

```python
print("Training size is:",X_train.shape,y_train.shape)
print("Testing size is:",X_test.shape,y_test.shape)
```

```python
X_train
```

```python
y_train=y_train.reset_index()
del y_train['index']
```

```python
y_test=y_test.reset_index()
del y_test['index']
```

```python
X_train=X_train.reset_index()
del X_train['index']
```

```python
X_test=X_test.reset_index()
del X_test['index']
```

```python
X_test
```

```python
Scaler=StandardScaler()
X_train_scaled=Scaler.fit_transform(X_train.drop('CHAS',axis=1))
X_test_scaled=Scaler.fit_transform(X_test.drop('CHAS',axis=1)) # standard scaling other indep variables except 'CHAS' which is categorical.
```

```python
X_train_scaled
```

```python
train_scaled_df=pd.DataFrame(X_train_scaled,columns=['CRIM', 'ZN', 'INDUS','NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT'])
test_scaled_df=pd.DataFrame(X_test_scaled,columns=['CRIM', 'ZN', 'INDUS','NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT'])
```

```python
train_scaled_df
```

```python
train_scaled_df['CHAS']=X_train.CHAS
test_scaled_df['CHAS']=X_test.CHAS
```

```python
LR.fit(train_scaled_df,y_train)
```

```python
C=B.drop('Predict',axis=1)
C_norm=C/C.max()
C_norm['Predict']=y_pred.flatten()
C_norm
```

```python
A=pd.DataFrame({'LSTAT':np.array(X_test.iloc[:,-1]),Housing Price(MEDV)':y_pred.flatten()})
A
```

```python
B.columns
```

```python
from scipy.stats import linregress
```

```python
text_y_start=0.5
text_y_delta=0.05

fig=plt.figure(figsize=(12,9))
features = ['CRIM', 'ZN', 'INDUS','NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']
colors=['red','green','blue','black','cyan','gray','navy','hotpink','olive','yellow','peachpuff','crimson']
line_styles=['-','--','-.',':','-','--','-.',':','-','--','-.',':']

for i,(features,color,line_style) in enumerate(zip(features,colors,line_styles)):
    sns.regplot(x=features,y='Predict',data=C_norm, color=color,label=features,scatter_kws={'s':5},line_kws={'linestyle':line_style})
    slope,intercept,r_value,p_value,std_err=linregress(C_norm[features],B['Predict'])
    equation_text=f'{features}:y={slope:.2f}x+{intercept:.2f} (r={r_value:.2f})'

    plt.text(0.05,text_y_start-i*text_y_delta,equation_text,transform=plt.gca().transAxes,color=color)

plt.legend(title='features')
plt.title('Linear Regression:features vs Housing Price')
plt.xlabel('Features')
plt.ylabel('Housing Price(MEDV)')
plt.savefig('E:/Analysis,Visulaization/HW3/LRH.png');
```

The above graph shows that the r-value is higher for feature ‘LSTAT’ and ‘RM’.
‘LSTAT’ has negative relation with ‘Housing Price’. We can say that housing price decreases with increase in value of
‘ % lower status population’. i.e, with increase in lower status population, housing price decreases that makes sense. 
Similarly the housing price increases with increase in RM , i.e, with the increase in room per dwelling, hosing price increases.
For other features, the relation is moderate with housing price.

```python
num_cols = 3  
num_rows = (len(B.columns) - 1) // num_cols + 1  


fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5*num_rows))

axes = axes.flatten()


for i, feature in enumerate(B.columns[:-1]):
    sns.regplot(x=feature, y='Predict', data=B, ax=axes[i])
    axes[i].set_title(f'Regression plot for {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Housing Price(MEDV)')


for j in range(i+1, num_cols*num_rows):
    fig.delaxes(axes[j])

plt.tight_layout();
```
